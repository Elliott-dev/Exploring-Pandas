```python
# Import dependencies
import pandas as pd
import os
import csv
```

```python
# Read file into DataFrame
Coffeeshop_reviews  = os.path.join('../','Resources', 'ratings_and_sentiments.csv')
```

```python
Coffeeshop_reviews_df = pd.read_csv(Coffeeshop_reviews)
```

```python
Coffeeshop_reviews_df 
```

```python
# Create a new DataFrame with the following columns:
# "coffee_shop_name", "review_text", "num_rating", and "overall_sent"
new_df=Coffeeshop_reviews_df[["coffee_shop_name", "review_text", "num_rating", "overall_sent"]]
```

```python
new_df
```

```python
# Find all reviews with a number rating greater than or equal to 4.
reviews_greater_equal_4=new_df.loc[new_df['num_rating'] >=4, 'coffee_shop_name']
a=len(reviews_greater_equal_4)
a
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
sentiment_greater_equal_4=new_df.loc[new_df['overall_sent'] >=4, 'coffee_shop_name']
a_1=len(sentiment_greater_equal_4)
a_1
```

```python
# Find the number of coffee shops with an overall sentiment greater than or equal to 4 *and* an overall rating greater than 4.
sentiment_greater_equal_4_w_reviews_greater_equal_4 = new_df[(new_df['overall_sent'] >=4) & (new_df['num_rating'] >=4)]
sentiment_greater_equal_4_w_reviews_greater_equal_4
b=len(sentiment_greater_equal_4_w_reviews_greater_equal_4)
b
```

```python
# Change the name of the review_text column to comments, and overall_sent to sentiment_rating.
new_df=new_df.rename(columns={"review_text":"comments", "overall_sent":"sentiment_rating"})
new_df
```

```python
# Save your changes to the file as a pipe delimited file.
new_df.to_csv('coffeeShop_reviews_output.csv', sep = '|')
```
